```python
import pandas as pd
import numpy as np

all_files=[]
import os
for root, dirs, files in os.walk("./dataset/multivariate_point_datasets"):
    for file in files:
        if file.endswith(".csv") or file.endswith(".txt"):
             all_files.append(os.path.join(root, file))
                
all_files=[dataset.replace("./dataset/","") for dataset in all_files]
df =pd.DataFrame({"Dataset":all_files})
Dataset_length=[]
nb_discords=[]
types=[]
position_discords=[]
for idx in range(0,len(df)):
    print(df['Dataset'][idx])
    
    if "SKAB" in df['Dataset'][idx] :
        dataset =pd.read_csv("dataset/"+df['Dataset'][idx],sep=";")
        types.append(df['Dataset'][idx].replace("multivariate_point_datasets/","").split("/")[0].split(".")[0])
        Dataset_length.append(len(dataset))

        anomalies=dataset["changepoint"]
        dataset_clean=dataset[dataset.columns[1:-2]]
        directory = os.path.dirname("dataset/clean_multivariate/"+df['Dataset'][idx].replace("multivariate_point_datasets/",""))
        if not os.path.exists(directory):
            os.makedirs(directory)
        dataset_clean.to_csv("dataset/clean_multivariate/"+df['Dataset'][idx].replace("multivariate_point_datasets/",""),index=False)
        
        all_files[idx]="clean_multivariate/"+df['Dataset'][idx].replace("multivariate_point_datasets/","")
        nb_discords.append(len(np.where(anomalies!=0.0)[0]))
        position_discords.append(str(list(np.where(anomalies!=0.0))).replace("[array([","").replace("])]","").replace(",",";"))
        
    if "creditcard" in df['Dataset'][idx]:
        dataset =pd.read_csv("dataset/"+df['Dataset'][idx],sep=",")
        types.append(df['Dataset'][idx].replace("multivariate_point_datasets/","").split("/")[0].split(".")[0])
        Dataset_length.append(len(dataset))
        
        anomalies=dataset["Class"]
        dataset_clean=dataset[dataset.columns[1:-1]]
        directory = os.path.dirname("dataset/clean_multivariate/"+df['Dataset'][idx].replace("multivariate_point_datasets/",""))
        if not os.path.exists(directory):
            os.makedirs(directory)
        dataset_clean.to_csv("dataset/clean_multivariate/"+df['Dataset'][idx].replace("multivariate_point_datasets/",""),index=False)
        
        all_files[idx]="clean_multivariate/"+df['Dataset'][idx].replace("multivariate_point_datasets/","")
        nb_discords.append(len(np.where(anomalies!=0)[0]))
        
        position_discords.append(str(list(np.where(anomalies!=0))).replace("[array([","").replace("])]","").replace(",",";"))
    
    if "kdd" in df['Dataset'][idx]:
        dataset=pd.read_csv("dataset/"+df['Dataset'][idx],sep=",",header=None,names=['duration','protocol_type','service','flag','src_bytes','dst_bytes','land','wrong_fragment','urgent','hot','num_failed_logins','logged_in','num_compromised','root_shell','su_attempted','num_root','num_file_creations','num_shells','num_access_files','num_outbound_cmds','is_host_login','is_guest_login','count','srv_count','serror_rate','srv_serror_rate','rerror_rate','srv_rerror_rate','same_srv_rate','diff_srv_rate','srv_diff_host_rate','dst_host_count','dst_host_srv_count','dst_host_same_srv_rate','dst_host_diff_srv_rate','dst_host_same_src_port_rate','dst_host_srv_diff_host_rate','dst_host_serror_rate','dst_host_srv_serror_rate','dst_host_rerror_rate','dst_host_srv_rerror_rate','Class'])
        types.append(df['Dataset'][idx].replace("multivariate_point_datasets/","").split("/")[0].split(".")[0])
        anomalies=dataset["Class"]
        Dataset_length.append(len(dataset))
        
        dataset_clean=dataset[dataset.columns[:-1]]
        directory = os.path.dirname("dataset/clean_multivariate/"+df['Dataset'][idx].replace("multivariate_point_datasets/",""))
        if not os.path.exists(directory):
            os.makedirs(directory)
        dataset_clean.to_csv("dataset/clean_multivariate/"+df['Dataset'][idx].replace("multivariate_point_datasets/","").replace(".txt",".csv"),index=False)
        
        all_files[idx]="clean_multivariate/"+df['Dataset'][idx].replace("multivariate_point_datasets/","").replace(".txt",".csv")
        nb_discords.append(len(np.where(anomalies!="normal.")[0]))
        print(anomalies[:10].values, len(np.where("normal." in anomalies.values)[0]))
        position_discords.append(str(list(np.where(anomalies!="normal."))).replace("[array([","").replace("])]","").replace(",",";"))


print(all_files)
df["Dataset"]=all_files
df["Dataset length"]=Dataset_length
df["Type"]=types
df["#discords"]=nb_discords
df["Position discord"]=position_discords
df["discord length"]=200
df.to_csv("multivariate_abnormal_point.csv",index=False)
df
```
